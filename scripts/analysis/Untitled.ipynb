```python
import sys
import json
import os

import sage_data_client

def get_events(vsn):
    return sage_data_client.query(
    start="-1h",
    filter={
        "name": "sys.scheduler.status.plugin.*",
        "vsn": vsn,
        "plugin": "scheduler:0.9.6",}
    )

def reconstruct_events(df):
    for _, i in df.iterrows():
        v = json.loads(i.value)
        v.update({"ts": i.timestamp, "event": i["name"]})
        yield v
```

```python
vsn = "W044"
df = get_events(vsn)

# Filter only events related to plugin execution
launched_plugins = df[df.name.str.contains("launched")]
complete_plugins = df[df.name.str.contains("complete")]
failed_plugins = df[df.name.str.contains("failed")]
```
